```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
#The below files contains the data behind "Do Voters Want Democrats or Republicans in Congress?"
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/congress-generic-ballot/generic_topline_historical.csv')
df1 = pd.read_csv("generic_ballot_averages18.csv")
```

```python
#maximum and minimum values of democratic and republic polls
print("The highest democratic polls :", df["dem_hi"].max())
print("The lowest democratic polls :", df["dem_lo"].min())
print("The highest republicans polls :",df["rep_hi"].max())
print("The lowest republicans polls :", df["rep_lo"].min())
```

```python
#Seperating democrats and republicans from given data.
de = df1.loc[df1["candidate"] =='Democrats'].reset_index() #resetting index
re = df1.loc[df1["candidate"] =='Republicans'].reset_index() #resetting index
```

```python
#plotting graph between democrats and republicans
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(2)

plt.plot(de["pct_estimate"],label = "Democrats")
plt.plot(re["pct_estimate"],label = "Republicans")
plt.title("Democratics vs Republicans ")
plt.ylabel("percent of votes")
plt.xlabel("Data points from ( 2018 â€“ 2022)")
plt.grid()
plt.legend()
plt.show()
```

```python
#getting dem_estimate and rep_estimate columns to draw graph.
dem = df["dem_estimate"]
rep = df["rep_estimate"]
```

```python
#plotting graph betweend highest and lowest votes for democrats and republicans over years
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(6)

plt.plot(df["rep_hi"],label = "rep_hi")
plt.plot(df["dem_hi"],label = "dem_hi")
plt.plot(df["rep_lo"],label = "rep_lo")
plt.plot(df["dem_lo"],label = "dem_lo")

plt.title("Democratics and republicans highest vs lowest data")
plt.ylabel("Percentage")
plt.xlabel("Data over years from 1995-2018 ")
plt.grid()
plt.legend()
plt.show()
```

```python
#histogram between democrats and republicans

f = plt.figure()
f.set_figwidth(10)
f.set_figheight(6)

plt.hist(rep,alpha = 0.7, label = "republicans")
plt.hist(dem, alpha = 0.7,label = "democrats")
plt.title("Comparsions of Democratics and republicans")
plt.xlabel("percentage of votes")
plt.ylabel("data over Years")
plt.show()
```

```python
#plotting pie chart for democrats and republicans
c = df[['dem_estimate','rep_estimate']].mean()

plt.pie(c,labels=["dem_estimate",'rep_estimate'],autopct='%1.1f%%')
plt.title("Overall percentage of votes for Democratics and Republicans")
plt.show()
```
